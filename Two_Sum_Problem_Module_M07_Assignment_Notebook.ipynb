## Performance Comparison: Brute Force vs Hash Map

**Problem:** Given an array of integers and a target value, return the indices of two numbers that add up to the target.

**Your Task:**

2. Write code to compare time and space usage of both approaches

## Import Libraries and Generate Test Data

```python
import time
import random
import sys

# Generate test arrays of different sizes
array_sizes = [10000, 50000, 100000]

test_arrays = {}
for size in array_sizes:
    test_arrays[size] = [random.randint(1, size * 10) for _ in range(size)]
    print(f"Generated array of size {size:,}")

# Set a target value to search for
target = 5000
print(f"\nTarget value: {target}")
```

## Hash Map Solution (Provided)

This is the hash map solution from the module.

```python
def two_sum_hash_map(nums, target):
    """
    Hash Map approach: Use dictionary for fast lookups
    """
    # Create an empty dictionary to store numbers we've seen
    seen = {}

    # Loop through each number with its index
    for i in range(len(nums)):
        current_number = nums[i]
        complement = target - current_number

        # Check if the complement exists in our dictionary
        if complement in seen:
            # Found it! Return the two indices
            return [seen[complement], i]

        # Store this number and its index in the dictionary
        seen[current_number] = i

    # No solution found
    return None

# Test it
test = [2, 7, 11, 15]
result = two_sum_hash_map(test, 9)
print(f"Test: {test}, target = 9")
print(f"Result: {result}")
```

## YOUR TASK 1: Implement Brute Force Solution

Write the brute force solution using nested loops to check all possible pairs.

```python
def two_sum_brute_force(nums, target):
    """
    Brute Force approach: Check all possible pairs
    """
    # TODO: Write your code here
    # Use two nested loops:
    # - Outer loop: for i in range(len(nums)):
    for i in range(len(nums)):
    # - Inner loop: for j in range(i + 1, len(nums)):
        for j in range(i + 1, len(nums)):
    # - Check if nums[i] + nums[j] == target
            if nums[i] + nums[j] == target:
    # - If yes, return [i, j]
                return [i, j]

    # No solution found
    return None

# Test case
test = [2, 7, 11, 15]
result = two_sum_brute_force(test, 9)
print(f"Test: {test}, target = 9")
print(f"Result: {result}")
```
